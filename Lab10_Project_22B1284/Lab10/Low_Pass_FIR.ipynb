```python
import numpy as np
from scipy.signal import firwin
import matplotlib.pyplot as plt
```

```python
cutoff_frequency = 2000  # in Hz
sampling_frequency = 20000  # in Hz
filter_taps = 8

nyquist_frequency = sampling_frequency / 2
normalized_cutoff_frequency = cutoff_frequency / nyquist_frequency

filter_coefficients = firwin(filter_taps, normalized_cutoff_frequency)
```

```python
print(filter_coefficients)
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Loading data from text file
file_path = "lpf_data.txt"
data = np.loadtxt(file_path)

#sampling interval
sampling_interval = 0.001041667 #9600 baud rate and 10 bits per data point

#time axis based on the sampling interval
time = np.arange(len(data)) * sampling_interval

# Plot the data against time
plt.plot(time, data)
plt.title('Plot of Data over Time with Regular Intervals')
plt.xlabel('Time')
plt.ylabel('Data Value')
plt.grid(True)
plt.show()

```
